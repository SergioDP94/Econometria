```python
import wooldridge as woo
import pandas as pd
import numpy as np
```

```python
# Cargar la base de datos MEAP01
data = woo.data('meap01')

# i) Valores mayor y menor de math4
math4_min = data['math4'].min()
math4_max = data['math4'].max()

print(f"Valor menor de math4: {math4_min}")
print(f"Valor mayor de math4: {math4_max}")
```

```python
# ii) Escuelas con tasa perfecta de aprobados en matemticas
perfect_scores = (data['math4'] == 100).sum()
percent_perfect = (perfect_scores / len(data)) * 100

print(f"Escuelas con tasa perfecta de aprobados en matemticas: {perfect_scores}")
print(f"Porcentaje del total: {percent_perfect:.2f}%")
```

```python
# iii) Escuelas con tasa de aprobados exactamente 50%
half_scores = (data['math4'] == 50).sum()
print(f"Escuelas con tasa de aprobados exactamente 50%: {half_scores}")
```

```python
# iv) Comparacin de promedios de math4 y read4
math4_mean = data['math4'].mean()
read4_mean = data['read4'].mean()

print(f"Promedio de aprobados en matemticas: {math4_mean:.2f}%")
print(f"Promedio de aprobados en lectura: {read4_mean:.2f}%")

if math4_mean < read4_mean:
    print("La prueba de matemticas parece ser ms difcil de aprobar.")
else:
    print("La prueba de lectura parece ser ms difcil de aprobar.")
```

```python
# v) Correlacin entre math4 y read4
correlation = data[['math4', 'read4']].corr().iloc[0, 1]
print(f"Correlacin entre math4 y read4: {correlation:.2f}")
```

```python
# vi) Promedio y desviacin estndar de exppp (gasto por alumno)
exppp_mean = data['exppp'].mean()
exppp_std = data['exppp'].std()

print(f"Gasto promedio por alumno: ${exppp_mean:.2f}")
print(f"Desviacin estndar del gasto por alumno: ${exppp_std:.2f}")
```

Hay una gran variacin en el gasto por alumno.

```python
# vii) Comparacin del gasto entre dos escuelas
gasto_a = 6000
gasto_b = 5000
porcentaje_superior = ((gasto_a - gasto_b) / gasto_b) * 100
log_diferencia = 100 * (np.log(gasto_a) - np.log(gasto_b))

print(f"El gasto de la escuela A supera al de la escuela B en un {porcentaje_superior:.2f}%.")
print(f"Diferencia porcentual aproximada basada en logaritmos: {log_diferencia:.2f}%.")
```
