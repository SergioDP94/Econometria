```python
import wooldridge as woo
import pandas as pd
import numpy as np
```

```python
# Cargar la base de datos JTRAIN2
data = woo.data('jtrain2')

# i) Proporci√≥n de hombres que recibieron capacitaci√≥n (train == 1)
train_prop = data['train'].mean()
print(f"Proporci√≥n de hombres que recibieron capacitaci√≥n laboral: {train_prop:.2%}")
```

```python
# ii) Promedio de ingresos en 1978 (re78) seg√∫n capacitaci√≥n
re78_train = data.loc[data['train'] == 1, 're78'].mean()
re78_no_train = data.loc[data['train'] == 0, 're78'].mean()

print(f"Ingresos promedio en 1978 para hombres con capacitaci√≥n: ${re78_train:.2f}")
print(f"Ingresos promedio en 1978 para hombres sin capacitaci√≥n: ${re78_no_train:.2f}")
```

```python
# Diferencia econ√≥mica
diff_income = re78_train - re78_no_train
print(f"Diferencia econ√≥mica en ingresos promedio: ${diff_income:.2f}")
```

```python
# iii) Proporci√≥n desempleada en 1978 seg√∫n capacitaci√≥n
unem_train = data.loc[data['train'] == 1, 'unem78'].mean()
unem_no_train = data.loc[data['train'] == 0, 'unem78'].mean()
print(f"Proporci√≥n desempleada en 1978 para hombres con capacitaci√≥n: {unem_train:.2%}")
print(f"Proporci√≥n desempleada en 1978 para hombres sin capacitaci√≥n: {unem_no_train:.2%}")
```

```python
# iv) Evaluaci√≥n de la efectividad del programa
print("El programa parece haber tenido un impacto positivo en los ingresos y el desempleo.")
```
