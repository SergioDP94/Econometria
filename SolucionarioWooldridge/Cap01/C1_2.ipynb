```python
import wooldridge as woo
import pandas as pd
```

```python
# Cargar la base de datos BWGHT
data = woo.data('bwght')

# i) Número de mujeres y mujeres que fumaron durante el embarazo
num_mujeres = len(data)
fumaron = (data['cigs'] > 0).sum()

print(f"Total de mujeres en la muestra: {num_mujeres}")
print(f"Mujeres que fumaron durante el embarazo: {fumaron}")
```

```python
# ii) Cantidad promedio de cigarros consumidos por día
cigs_mean = data['cigs'].mean()
print(f"Promedio de cigarros consumidos por día: {cigs_mean:.2f}")
```

```python
# iii) Promedio de cigarros entre mujeres que fumaron
cigs_fumadoras = data.loc[data['cigs'] > 0, 'cigs'].mean()
print(f"Promedio de cigarros entre mujeres que fumaron: {cigs_fumadoras:.2f}")
```

```python
# iv) Promedio de fatheduc (años de educación del padre) y número de observaciones válidas
fatheduc_mean = data['fatheduc'].mean()
observaciones_validas = data['fatheduc'].count()

print(f"Años promedio de educación del padre: {fatheduc_mean:.2f}")
print(f"Observaciones válidas para calcular fatheduc: {observaciones_validas}")
```

```python
# v) Promedio e desviación estándar del ingreso familiar (famincome)
famincome_mean = data['faminc'].mean()
famincome_std = data['faminc'].std()

print(f"Ingreso familiar promedio: ${famincome_mean:.2f}")
print(f"Desviación estándar del ingreso familiar: ${famincome_std:.2f}")
```
