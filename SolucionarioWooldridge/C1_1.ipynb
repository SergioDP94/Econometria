```python
import wooldridge as woo
import pandas as pd
```

```python
# Cargar la base de datos WAGE1
data = woo.data('wage1')

# i) Nivel educativo promedio, menor y mayor
education_mean = data['educ'].mean()
education_min = data['educ'].min()
education_max = data['educ'].max()

print(f"Nivel educativo promedio: {education_mean:.2f}")
print(f"Nivel educativo menor: {education_min}")
print(f"Nivel educativo mayor: {education_max}")
```

```python
# ii) Salario promedio por hora
wage_mean = data['wage'].mean()
print(f"Salario promedio por hora en 1976: ${wage_mean:.2f}")
```

```python
# iii) Índices de precios al consumidor (CPI) para 1976 y 2003
cpi_1976 = 56.9  # Ejemplo: CPI 1976
cpi_2003 = 184.0  # Ejemplo: CPI 2003

# iv) Ajustar salario promedio a dólares de 2003
wage_adjusted = wage_mean * (cpi_2003 / cpi_1976)
print(f"Salario promedio por hora ajustado a dólares de 2003: ${wage_adjusted:.2f}")
```

```python
# v) Número de mujeres y hombres
female_count = data['female'].sum()
male_count = len(data) - female_count

print(f"Número de mujeres en la muestra: {female_count}")
print(f"Número de hombres en la muestra: {male_count}")
```
